```python
metric_list = ['valid_AUC','valid_AUC_PR','valid_Recall','valid_F1',
               
               'valid_AUC_tuned','valid_AUC_PR_tuned','valid_Recall_tuned','valid_F1_tuned',
               
              'test_AUC','test_AUC_PR','test_Recall','test_F1',
               
               'test_AUC_tuned','test_AUC_PR_tuned','test_Recall_tuned','test_F1_tuned']
```

```python
import pandas as pd
T5 = pd.DataFrame(columns =metric_list, index = selected_assay_list)
```

```python
base_ismorehighthan_tuned1 = []
base_ismorehighthan_tuned2 = []
for a in selected_assay_list : 
    result_df = pd.read_csv('./31assays/performance/performance'+a+'.csv').iloc[:,1:]
    
    
    #if base_f1 > tuned_f1 
    if result_df.iloc[0,2] >= result_df.iloc[1,2] : 
        result_df.iloc[1,:] = result_df.iloc[0,:]
        base_ismorehighthan_tuned1.append(a)
    if result_df.iloc[2,2] >= result_df.iloc[3,2] : 
        result_df.iloc[3,:] = result_df.iloc[2,:]    
        base_ismorehighthan_tuned2.append(a)

#     TF_index = result_df.iloc[0,:] > result_df.iloc[1,:]
#     result_df.iloc[1,:][TF_index] = result_df.iloc[0,:]
    
#     TF_index = result_df.iloc[2,:] > result_df.iloc[3,:]
#     result_df.iloc[3,:][TF_index] = result_df.iloc[2,:]
    
    T5.loc[a,:] = result_df.to_numpy().flatten()
```

```python
set1 = set(base_ismorehighthan_tuned1 )
set2 = set(base_ismorehighthan_tuned2 )
set3 =(set1.intersection(set2))
```
