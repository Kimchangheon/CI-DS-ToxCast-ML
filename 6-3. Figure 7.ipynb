```python
import matplotlib.pyplot as plt
import numpy as np


plt.rc('font', size=15) 
plt.rc('font', weight = 'bold')

def main():
	#### 1. bar plot   
	models = ['Validation F1', 'Test F1(Within AD)']
	xticks = ['NVS_ADME_hCYP2B6', 'NVS_ADME_hCYP1A2', 'NCCT_TPO_AUR_dn']
	data = {
			'Validation F1':T5_within.iloc[28:,:]['valid_F1_tuned'],
            'Test F1(Within AD)' : T5_within.iloc[28:,:]['test_F1_tuned']
           }
	
	#### 2. matplotlib figure  axis 
	fig, ax = plt.subplots(1,1,figsize=(7,5)) # 1x1 figure matrix , (7)x(5) 
	colors = ['#B689C0','#EEF3D2']
	width = 0.15
	
	#### 3. bar 
	for i, model in enumerate(models):
		pos = compute_pos(xticks, width, i, models)
		bar = ax.bar(pos, data[model], width=width*0.95, label=model, color=colors[i])
		present_height(ax, bar) # bar 
	
	#### 4. x 
	ax.set_xticks(range(len(xticks)))
	ax.set_xticklabels(xticks, fontsize=12)	
	# ax.set_xlabel('Cancer type', fontsize=14)
	
	#### 5. y 
	ax.set_ylim([0.5,0.76])
	ax.set_yticks([0.0, 0.40, 0.60, 0.65, 0.7,0.75, 0.8,1.0])
	ax.yaxis.set_tick_params(labelsize=10)
	# ax.set_ylabel('Prediction Accuracy', fontsize=14)
	
	#### 6.  
	ax.legend(loc='upper right', shadow=True, ncol=1)
	
	#### 7. () 
	ax.set_axisbelow(True)
	ax.yaxis.grid(True, color='gray', linestyle='dashed', linewidth=0.5)
	
	#### 8.   
	plt.tight_layout()
	plt.savefig('T6.png', format='png', dpi=300)
	plt.show()
	
def compute_pos(xticks, width, i, models) :
  index = np.arange(len(xticks))
  n = len(models)
  correction = i-0.5*(n-1)
  return index + width*correction

def present_height(ax, bar):
	for rect in bar:
		height = rect.get_height()
		posx = rect.get_x()+rect.get_width()*0.5
		posy = height*0.5
		ax.text(posx, posy, '%.3f' % height, rotation=90, ha='center', va='bottom')
		
if __name__=='__main__':
	main()
```
