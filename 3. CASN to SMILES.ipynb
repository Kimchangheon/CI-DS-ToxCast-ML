The SMILES downloaded with this selection are for the chemical structure associated with the substance displayed in a particular record. They can be considered to be "substance-level SMILES". Other types of SMILES are available including "QSAR-ready" and "MS-Ready" forms. QSAR-Ready SMILES are those that were used in order to perform the OPERA predictions (processed to desalt, remove stereo and split multi-component chemicals into individual components). MS-Ready SMILES has similar processing but uses a different list of salt components to retain mappings.

MS-Ready SMILES are the SMILES representations of the desalted, de-isotoped, stereo-neutral forms of chemical structures associated with particular chemical substances. One MS-Ready SMILES can map to multiple individual chemical substances.

QSAR-Ready SMILES are the SMILES representations of the desalted, de-isotoped, stereo-neutral forms of chemical structures associated with particular chemical substances. One QSAR-Ready SMILES can map to multiple individual chemical substances. They differ from MS-Ready SMILES based on a more restricted set of salt counterions primarily.

```python
searched_df = pd.read_csv(path + assay_name+".csv") # engine="python", encoding = "utf-8" 하면 .iterrows()에서 i가 for문의 index가 아니라.. row가 되버림..
```

```python
#1-1) EPA comptox dashboard에서 batch단위로 smiles 찾아서 채우기
smiles_searched_1= {}#찾은 smiles 딕셔러니
remove_index = [] #CASN_nosmiles, CASN_nosmiles_index에서 찾은것들은 제외하기 위한 인덱스
for i, row in searched_df.iterrows() : 
  if row[-1] != '-' : 
    smiles_searched_1[i] = row['SMILES'] #qsar ready-smiles로 한다.
    remove_index.append(i)
count_nan = count_nan - len(smiles_searched_1)
print("count_nan : ",count_nan)
#찾은 smiles 채우기
for key in smiles_searched_1.keys() : 
  smiles_list[ CASN_nosmiles_index[key] ] = smiles_searched_1[key]
#CASN_nosmiles과 CASN_nosmiles_index에서 찾은 것들은 지우기
CASN_nosmiles_2 = []
CASN_nosmiles_index_2 = []
remove_index_list = smiles_searched_1.keys()
for i in range(0,len(CASN_nosmiles)) : 
  if i in remove_index_list : 
    pass
  else : 
    CASN_nosmiles_2.append(CASN_nosmiles[i])
    CASN_nosmiles_index_2.append(CASN_nosmiles_index[i])
print( pd.Series( smiles_list).isnull().sum() )
print(len(CASN_nosmiles_2))
print(len(CASN_nosmiles_index_2))


# 1-2) PubchemPy패키지 이용하여 채우기
import pubchempy as pcp

searched_df_2 = [] 

for casn in CASN_nosmiles_2 : 
  # cmpnd = pcp.get_compounds(casn, 'name') #PubChem ID cid가 출력된다. 
  try : 
    cmpnd = pcp.get_compounds(casn, 'name') #PubChem ID cid가 출력된다. \
    searched_df_2.append(cmpnd[0].canonical_smiles)
  except : 
    searched_df_2.append('-')

smiles_searched_2 ={}
remove_index = [] #CASN_nosmiles_2, CASN_nosmiles_index_2에서 찾은것들은 제외하기 위한 인덱스
for i, row in enumerate(searched_df_2) : 
  if row != '-' : 
    smiles_searched_2[i] = row
    remove_index.append(i)
count_nan = count_nan - len(smiles_searched_2)
print("count_nan : ", count_nan)
#찾은 smiles채우기
for key in smiles_searched_2.keys() : 
  smiles_list[ CASN_nosmiles_index_2[key] ] = smiles_searched_2[key]
#CASN_nosmiles_2과 CASN_nosmiles_index_2에서 찾은 것들은 지우기
CASN_nosmiles_3 = []
CASN_nosmiles_index_3 = []
remove_index_list = smiles_searched_2.keys()
for i in range(0,len(CASN_nosmiles_2)) : 
  if i in remove_index_list : 
    pass
  else : 
    CASN_nosmiles_3.append(CASN_nosmiles_2[i])
    CASN_nosmiles_index_3.append(CASN_nosmiles_index_2[i])
print(pd.Series(smiles_list).isnull().sum())
print(len(CASN_nosmiles_3))
print(len(CASN_nosmiles_index_3))

#1-3) CIRconvert메서드 이용하여 채우기
from urllib.request import urlopen

def CIRconvert(ids):
    try:
        url = 'http://cactus.nci.nih.gov/chemical/structure/' + ids + '/smiles'
        ans = urlopen(url).read().decode('utf8')
        return ans
    except:
        return 'Did not work'


searched_df_3 = [] 

for casn in CASN_nosmiles_3 : 
  smiles = CIRconvert(casn)
  if smiles != 'Did not work' : 
      searched_df_3.append(smiles)
  else : 
      searched_df_3.append('-')

smiles_searched_3 ={}
remove_index = [] #CASN_nosmiles_3, CASN_nosmiles_index_3에서 찾은것들은 제외하기 위한 인덱스
for i, row in enumerate(searched_df_3) : 
  if row != '-' : 
    smiles_searched_3[i] = row
    remove_index.append(i)

count_nan = count_nan - len(smiles_searched_3)
print("count_nan : ",count_nan)

for key in smiles_searched_3.keys() : 
  smiles_list[ CASN_nosmiles_index_3[key] ] = smiles_searched_3[key]

CASN_nosmiles_4 = []
CASN_nosmiles_index_4 = []
remove_index_list = smiles_searched_3.keys()
for i in range(0,len(CASN_nosmiles_3)) : 
  if i in remove_index_list : 
    pass
  else : 
    CASN_nosmiles_4.append(CASN_nosmiles_3[i])
    CASN_nosmiles_index_4.append(CASN_nosmiles_index_3[i])

print(pd.Series(smiles_list).isnull().sum())
print(len(CASN_nosmiles_4))
print(len(CASN_nosmiles_index_4))

print("smiles를 찾지 못한 cans 리스트 : " ,CASN_nosmiles_4)
# df = df.to_frame()
df.insert(1, 'smiles', smiles_list)
df.to_csv(path + assay_name+"_df.csv")
```
