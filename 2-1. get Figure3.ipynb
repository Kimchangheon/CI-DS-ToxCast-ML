```python
import pandas as pd
T2= pd.read_csv("T2.csv")
percentage_active = []
for i in T2['No. of active chemicals (%)'] : 
    i2 = float(i.replace("%",""))
    percentage_active.append(i2/100)
```

```python
T2['actives'] = T2['No. of chemicals'] * percentage_active
T2['inactives'] = T2['No. of chemicals'] * ( pd.Series([1]*31) - pd.Series(percentage_active))
```

```python
T2.index = T2['Assay Name']
```

```python
T2 = T2.iloc[::-1]
```

```python
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Times New Roman'
mpl.rcParams['font.size'] = 10
```

```python
import matplotlib.pyplot as plt

fields = ['inactives','actives']
colors = ['#DBE2EF', '#3F72AF']
labels = ['inactives','actives']
# figure and axis
fig, ax = plt.subplots(1, figsize=(12, 10),dpi=1600)
# plot bars
left = len(T2) * [0]
for idx, name in enumerate(fields):
    plt.barh(T2.index, T2[name], left = left, color=colors[idx])
    left = left + T2[name]
# title, legend, labels
plt.title('Selected assays for class imbalance (with 7000 or more chemicals) and data scarcity (with inactive ratio of 40–60%)\n', loc='left')
plt.legend(labels, bbox_to_anchor=([0.55, 1, 0, 0]), ncol=4, frameon=False)
plt.xlabel('Number of chemicals')
# remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
# adjust limits and draw grid lines
plt.ylim(-0.5, ax.get_yticks()[-1] + 0.5)
ax.set_axisbelow(True)
ax.xaxis.grid(color='gray', linestyle='dashed')
plt.savefig('T2.png',dpi=1600)
plt.show()
```
