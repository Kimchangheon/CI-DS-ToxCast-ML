```python
for assay in selected_assay_list : 
    df = pd.DataFrame(columns =fp_list, index = list(map(lambda x : x+'_AUC', al_list))+
                      list(map(lambda x : x+'_Recall', al_list))
                      +list(map(lambda x : x+'_F1', al_list)) )
    df.columns.name = assay
    for fp in fp_list : 
        for al in al_list : 
            value = pd.read_csv("./result_sklearn_all_a_fp/metric/"+fp+"/"+al+"/"+assay+'_metric.csv'
                               , index_col=0)
            value = value.loc['Mean']
#             value = [value['AUC'],value['Recall'],value['F1']]
            for metric in ['AUC','Recall','F1'] : 
                df.loc[al+'_'+metric,fp] = value[metric]
            
    df.to_csv("./each_assay_result/"+assay+'.csv',index=True)
    

    
```
