```python
import pandas as pd
assays = pd.read_csv("31assays_name.csv")
assays = list(assays['Assay Name'])
```

```python
assays.remove("TOX21_NFkB_BLA_agonist_ratio")
```

```python
len(assays)
```

```python
i = "TOX21_VDR_BLA_agonist_ratio"
columns = pd.read_csv("./31assays/params/"+i+'.csv').columns
```

```python
columns
```

```python
parameters_table = pd.DataFrame(columns = columns)
for i in assays : 
    df = pd.read_csv("./31assays/params/"+i+'.csv')
    parameters_table = pd.concat([parameters_table,df])
```

```python
parameters_table.index = parameters_table['Unnamed: 0']
parameters_table = parameters_table.drop('Unnamed: 0', axis = 1)
parameters_table.index.name = 'assays'
parameters_table.columns = ['n_estimators','min_samples_split', 'min_samples_leaf',
                            'rf__max_features', 'rf__max_depth','rf__bootstrap']
```

```python
base_ismorehighthan_tuned = ['TOX21_ESRE_BLA_ratio',
 'TOX21_FXR_BLA_antagonist_ratio',
 'TOX21_VDR_BLA_agonist_ratio',
 'TOX21_PPARg_BLA_antagonist_ratio',
 'TOX21_TR_LUC_GH3_Agonist',
 'TOX21_PPARg_BLA_Agonist_ratio',
 'NVS_ADME_hCYP1A2',
 'TOX21_TR_LUC_GH3_Antagonist',
 'TOX21_p53_BLA_p2_ratio',
 'TOX21_AhR_LUC_Agonist']
```

```python
parameters_table
```

```python
import numpy as np
for i in base_ismorehighthan_tuned : 
    parameters_table.loc[i,:] = pd.Series({'n_estimators':100, 'min_samples_split':2, 'min_samples_leaf' : 1, 'rf__max_features': "sqrt", 
                                                   'rf__max_depth' : np.nan, 'rf__bootstrap' : True})
```
